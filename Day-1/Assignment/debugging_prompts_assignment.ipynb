# 📘 Assignment 2: Debugging Prompts


## 🧩 Objective
Improve poorly performing prompts by:
- Identifying issues
- Fixing prompts
- Comparing before vs after
- Explaining improvements



## 🧠 Introduction
Prompt engineering directly impacts AI response quality. Poor prompts lead to unclear, incorrect, or verbose outputs.

This notebook demonstrates how to debug prompts effectively.



## 🧱 Common Problems in Prompts
- Lack of context  
- Vague instructions  
- No structure  
- Missing role definition  
- No tone specification  



## ✅ Case 1: Order Status

### ❌ Before
Where is my order?

### 🔍 Issues
- No order ID  
- No structure  

### ✅ After
You are a customer support assistant.

Customer Query: "Where is my order #12345?"

Provide:
- Current status  
- Delivery estimate  
- Apology if delayed  

### 💡 Improvement
Added context, structure, and clarity.



## ✅ Case 2: Troubleshooting

### ❌ Before
My app is not working.

### 🔍 Issues
- Too vague  
- No steps  

### ✅ After
You are a technical support assistant.

Customer Issue: "My app is not working."

Provide:
- Step-by-step solution (max 5 steps)  
- Simple language  

### 💡 Improvement
Added structured steps and constraints.



## ✅ Case 3: Refund

### ❌ Before
I want a refund.

### 🔍 Issues
- No tone  
- No clarity  

### ✅ After
You are a support agent.

Customer Request: "I want a refund."

Provide:
- Confirmation  
- Processing time (5–7 days)  
- Polite tone  

### 💡 Improvement
Improved tone and clarity.



## ✅ Case 4: Login Issue

### ❌ Before
I can't log in.

### 🔍 Issues
- No solution steps  

### ✅ After
You are a technical support assistant.

Provide:
- Causes  
- Reset steps  
- Contact support option  

### 💡 Improvement
Added complete guidance.



## ✅ Case 5: Payment Failure

### ❌ Before
Payment failed. What to do?

### 🔍 Issues
- No explanation  
- No reassurance  

### ✅ After
You are a support assistant.

Provide:
- Reasons  
- Retry steps  
- Reassurance  

### 💡 Improvement
Improved completeness and user confidence.
